<span style='font-size:large; color:blue'>
The possible permutations of letters a , b and c
</span>

There is 6 possible permutation: **abc, bac, acb, bca, cab, cba**

```python
# using python's math library
math.perm(3,3)
# where n=3 and k=3 (number of items or r)
```

since **abc** and **cba** can be applied (order matters); $n!$ is the easy calculation.

```python
# n!
math.factorial(3)
```

```python
# print all possibilities with itertools
import itertools
for c, i in enumerate(itertools.permutations(['a','b','c'])):
    print(c,i)
```

```python
# scipy: permutation
scipy.special.perm(3,3)
```

```python
# combinations of two pairs without order 
for c, i in enumerate(itertools.combinations(['a','b','c'],2)):
    print(c,i)
```

```python
# scipy: combination without repetition
scipy.special.comb(N=3, k=2, repetition=False, exact=True)
```

```python
# combination with repetition
for c, i in enumerate(itertools.combinations_with_replacement(['a','b','c'],2)):
    print(c,i)
```

```python
# scipy: combination with repetition
scipy.special.comb(N=3, k=2, repetition=True, exact=True)
```

<span style="font-size:large; color:blue">
Permutation: ice cream
</span>

I have 4 scoops of ice cream: 1 chocolate, 1 strawberry, 1 vanilla, and 1 mint.  
I want to eat only 3 of the scoops.  
How many different ways can I eat 3 of the scoops if I consider both which scoops I eat and the order in which
I eat them?

This is a permutation question, since I care about the order in which I eat the scoops.  
There are 4 total scoops, but I only want to eat 3 of them.

```python
# permutation
math.perm(4,3)
```

There are 24 different ways that I could eat 3 of the 4 scoops.  
For example, chocolate-strawberry-vanilla would be 1 of the 24 options,  
but since order matters, chocolate-vanilla-strawberry would be another option.

.

<span style="font-size:large; color:blue">
Combination: ice cream
</span>

I have the same 4 scoops of ice cream: 1 chocolate, 1 strawberry, 1 vanilla, and 1 mint.  
I want to eat only 3 of the scoops,  
and I donâ€™t care about the order in which I eat my 3 scoops.  
How many different combinations of 3 scoops can I create?

This is a combination question, since I donâ€™t care about the order in which I eat the scoops.  
There are 4 total scoops, but I only want to eat 3 of them.

```python
# combination
math.comb(4,3)
```

There are 4 different ways that I could eat 3 of the 4 scoops.  
For example, chocolate-strawberry-vanilla would be 1 of the 4 options,  
but since order doesnâ€™t matter,  
chocolate-vanilla-strawberry would be the same combination, and wouldnâ€™t count as another one of the 4 combinations.
